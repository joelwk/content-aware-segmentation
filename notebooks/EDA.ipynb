```python
import random
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
import math
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
# All .npy files in the embeddings directory

# Flatten the list of embeddings
embedding_values = np.vstack(embedding_values)

# Perform PCA and visualize
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(embedding_values)
plt.scatter(reduced_embeddings[:, 0], reduced_embeddings[:, 1], c=range(len(reduced_embeddings)))
plt.colorbar(label='Time (in seconds)')
plt.title('Frame Embeddings Over View Duration')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
