#### Get test videos
##### Run the cell below to get 5 random youtube videos and follow  the second data load option instructions.

````python
import pandas as pd

def convert_duration(duration_str):
    try:
        parts = duration_str.split(':')
        if len(parts) == 3:
            return int(parts[0]) * 3600 + int(parts[1]) * 60 + int(parts[2])
        elif len(parts) == 2:
            return int(parts[0]) * 60 + int(parts[1])
    except (AttributeError, ValueError):
        return 0

data = pd.read_parquet('./drive/MyDrive/emo-vlog-actors.parquet')
df = data.rename(columns={'link': 'url', 'title': 'caption'})

if 'duration' in df.columns:
    df['duration'] = df['duration'].apply(convert_duration)

def filter_videos_for_batch(video_df, min_duration_seconds=300, max_duration_seconds=3600, max_videos=50000):
    filtered_videos = []
    total_duration = 0
    for _, row in video_df.iterrows():
        video_duration = row['duration']
        if min_duration_seconds <= video_duration <= max_duration_seconds:
            filtered_videos.append(row)
            total_duration += video_duration
            if total_duration > max_duration_seconds or len(filtered_videos) >= max_videos:
                break
    return pd.DataFrame(filtered_videos)
_ = filter_videos_for_batch(df.sample(100), min_duration_seconds=int((3 * 60)), max_duration_seconds=int((30 * 60)), max_videos=5)
_.to_parquet('videos.parquet')
````
